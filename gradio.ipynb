```python
import gradio as gr
import numpy as np
from PIL import Image
from typing import Union

def predict(file: Union[str, bytes, None]) -> tuple[Image.Image, list[tuple[np.ndarray, str]]]:
    if file is None:
        raise gr.Error("No file provided.")

    # Load the input file as a grayscale image
    image = Image.open(file).convert("L")
    width, height = image.size

    # Create a dummy circular mask as a NumPy array (0-1 float values)
    mask = np.zeros((height, width), dtype=np.float32)
    # cx, cy, r = width // 2, height // 2, min(width, height) // 4
    # for y in range(height):
    #     for x in range(width):
    #         if (x - cx) ** 2 + (y - cy) ** 2 < r ** 2:
    #             mask[y, x] = 1.0  # Confidence 1.0 inside the circle

    # Return the image and a single annotation mask labeled "dummy"
    return image, [(mask, "dummy")]

demo = gr.Interface(
    fn=predict,
    inputs=gr.File(file_types=["image"], label="Upload Image File"),
    outputs=gr.AnnotatedImage(label="Prediction", color_map={"dummy": "#ff0000"})
)

if __name__ == "__main__":
    demo.launch()
```
